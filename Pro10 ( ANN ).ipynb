```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 784).astype('float32') / 255
x_test = x_test.reshape(10000, 784).astype('float32') / 255

y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
model_relu_adam = Sequential([
    Dense(512, activation='relu', input_shape=(784,)),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
])
model_relu_adam.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
model_relu_adam.fit(x_train, y_train, epochs=10)
test_loss, test_acc = model_relu_adam.evaluate(x_test, y_test)
print('Test accuracy with ReLU activation function and Adam optimizer:', test_acc)
model_sigmoid_sgd = Sequential([
    Dense(512, activation='sigmoid', input_shape=(784,)),
    Dense(256, activation='sigmoid'),
    Dense(10, activation='softmax')
])
model_sigmoid_sgd.compile(optimizer=SGD(lr=0.01),
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
model_sigmoid_sgd.fit(x_train, y_train, epochs=10)
test_loss, test_acc = model_sigmoid_sgd.evaluate(x_test, y_test)
print('Test accuracy with sigmoid activation function and SGD optimizer:', test_acc)
```
