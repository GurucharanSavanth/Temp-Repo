```python
#resize
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image = cv2.imread('/content/image.png')
cv2_imshow(image)
up_width = 800
up_height = 600
up_points = (up_width, up_height)
resized_up = cv2.resize(image, up_points)
cv2_imshow(resized_up)

#flip

import cv2 as cv
from google.colab.patches import cv2_imshow
img = cv.imread('image.png')
flipped_img=cv.flip(img, 1)
cv2_imshow(flipped_img)
cv2_imshow(img)
cv.waitKey(0)
cv.destroyAllWindows()
```
